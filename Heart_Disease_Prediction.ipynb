# Pre Processing the Dataset

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("heart_disease.csv")
df
```

```python
df.drop("education", axis=1, inplace=True)
```

```python
df
```

```python
df.dropna(inplace=True)
```

```python
df
```

```python
df["Gender"] = df["Gender"].map({"Male":1, "Female":0}).astype("int")
df["prevalentStroke"] = df["prevalentStroke"].map({"yes":1, "no":0}).astype("int")
```

```python
df
```

```python
df["Heart_ stroke"] = df["Heart_ stroke"].map({"yes":1, "No":0}).astype("int")
```

```python
df
```

# Store Feature Matrix in X and Traget in Vector y

```python
X = df.drop("Heart_ stroke", axis=1)
y = df["Heart_ stroke"]
```

```python
y.value_counts()
```

# Data Resampling

```python
pip install --upgrade scikit-learn imbalanced-learn
```

```python
from imblearn.over_sampling import SMOTE
```

```python
sm = SMOTE(random_state=2)
```

```python
X,y = sm.fit_resample(X, y.ravel())
```

```python
np.unique(y, return_counts=True)
```

# Model Creation

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
model = LogisticRegression()
```

```python
model.fit(X_train, y_train)
```

```python
model.score(X_test, y_test)
```

# Predicting Values from Model

```python
input = np.array([[0,59,0,0,0,0,1,0,209,150,85,20.77,90,88]])
```

```python
x = model.predict(input)
x[0]
```

```python
import pickle as pk
```

```python
pk.dump(model, open("Heart_disease_model.pkl", "wb"))
```
